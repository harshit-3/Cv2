```python
import cv2
import numpy as np
```

```python
#video capture 
#press 'q' to quit
cam = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi',fourcc,20.0, (640,480))

while(cam.isOpened()):
    ret, frame = cam.read()
    if ret == True:
        out.write(frame)
    
        grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow('frame', grey)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

cam.release()
out.release()
cv2.destroyAllWindows()
```

```python
#Draw shape on images
img = cv2.imread('images.jpg',1)
img = cv2.line(img,(0, 0),(255,255),(255,0 ,0),5)
img = cv2.circle(img, (342,432),74,(245,44,0),-1)
img = cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#to put text in live video
cam = cv2.VideoCapture(0)

while(cam.isOpened()):
    ret, frame = cam.read()
    if ret == True:
        text = ('harshit')
        font  = cv2.FONT_HERSHEY_SIMPLEX
        frame = cv2.putText(frame, text,(10,20),font, 1, (0,255,255),2, cv2.LINE_AA)
    
        cv2.imshow('frame', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

cam.release()
cv2.destroyAllWindows()
```

```python
#Right click to on image tp get to get color
#left click to get Position

def click_event(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        font = cv2.FONT_HERSHEY_PLAIN
        text = str(x) +' , ' + str(y)
        cv2.putText(img,text,(x,y),font, 1, (240,87,128), 2)
        cv2.imshow('image', img)
        
    if event == cv2.EVENT_RBUTTONDOWN:
        font = cv2.FONT_HERSHEY_PLAIN
        blue = img[y,x]
        green = img[y,x]
        red = img[y,x]
        text = str(blue)+' ' + str(green) + ' ' + str(red)
        cv2.putText(img, text, (x, y), font, 1, (255,0,0), 2)
        cv2.imshow('image', img)


img = cv2.imread('joker.jpg',0)
cv2.imshow('image',img)

cv2.setMouseCallback('image', click_event)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#object dection in image 
# to close 'esc'
cv2.namedWindow('Tracking')


def nothing(x):
    pass


cv2.createTrackbar('LH', 'Tracking', 0, 255, nothing)
cv2.createTrackbar('LS', 'Tracking', 0, 255, nothing)
cv2.createTrackbar('LV', 'Tracking', 0, 255, nothing)
cv2.createTrackbar('UH', 'Tracking', 255, 255, nothing)
cv2.createTrackbar('US', 'Tracking', 255, 255, nothing)
cv2.createTrackbar('UV', 'Tracking', 255, 255, nothing)



while True:
    img = cv2.VideoCapture(0)

    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    LH = cv2.getTrackbarPos('LH', 'Tracking')
    LS = cv2.getTrackbarPos('LS', 'Tracking')
    LV = cv2.getTrackbarPos('LV', 'Tracking')
    UH = cv2.getTrackbarPos('UH', 'Tracking')
    US = cv2.getTrackbarPos('US', 'Tracking')
    UV = cv2.getTrackbarPos('UV', 'Tracking')

    LB = np.array([LH, LS, LV])
    UB = np.array([UH, US, UV])
    mask = cv2.inRange(hsv, LB, UB)

    res = cv2.bitwise_and(img, img, mask=mask)


    cv2.imshow('image', img)
    cv2.imshow('res',res)
    cv2.imshow('mask',mask)
    key = cv2.waitKey(1)
    if key == 27:
        break


cv2.destroyAllWindows()
```
